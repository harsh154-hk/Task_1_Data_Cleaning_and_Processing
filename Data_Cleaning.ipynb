Data Cleaning And Processing 


Step 1 - Upload the dataset "Medical Appointment No Shows"

```python
import pandas as pd
```

```python
df=pd.read_csv("Medical Appointment No Shows.csv")
```

Step 2 - Now we will identify and manage the null values.

```python
df.info()
```

* As we can see that their is no null value in the Dataset

Step 3 - Manage the duplicate rows

```python
df.head(10)
```

* Now lets delete the duplicate rows on the basis of ApponmentID column...as it is unique 

```python
df.drop_duplicates(subset='AppointmentID', inplace=True)
```

```python
df.shape
```

* As there we can see there is no duplicate rows in the entire data set 

Step 4 - Standardize the text values

* Gender

```python
df['Gender'] = df['Gender'].str.upper().str.strip()
```

* Neighbourhood

```python
df['Neighbourhood'] = df['Neighbourhood'].str.upper().str.strip()
```

* No-Show

```python
df['No-show'] = df['No-show'].str.upper().str.strip() 
```

```python
df.head(5)
```

Step 5 - As we can see that we dont need ScheduledDay so we will drop this column from the data set

```python
df.drop(columns='ScheduledDay', inplace=True)
```

```python
df.head(5)
```

Step 6 - Correcting the date format

```python
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay']).dt.date
```

```python
df.head(5)
```

Step 7 - Rename column headers to be clean and uniform

```python
df.columns = (
    df.columns.str.strip()
              .str.lower()
              .str.replace(' ', '_')
              .str.replace('-', '_'))
```

```python
df.rename(columns={
    'hipertension': 'hypertension',
    'handcap': 'handicap'
    'patientid': 'patient_'
}, inplace=True)
```
